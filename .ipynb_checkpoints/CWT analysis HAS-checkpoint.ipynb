## Raw data

```python
len(data.index)
```

```python
sen4 = data['PRESSURE    AXIS    RANGE  0- 0.1 BAR      - Plot 4']
time = data.index
```

```python
fig, ax = plt.subplots(4, 2, figsize=(16, 16))

time_slices = [(i, i+20000) for i in range(0, 160000, 20000)]
for i, (start, end) in enumerate(time_slices):
    row = i // 2
    col = i % 2
    ax[row, col].plot(time[start:end], sen4[start:end])
    ax[row, col].set_title(f'{start} - {end} millisec')

plt.tight_layout()
plt.show()
```

## Smoothing Signal

```python
sen4_smooth = signal.savgol_filter(data['PRESSURE    AXIS    RANGE  0- 0.1 BAR      - Plot 4'],51,2)
time = data.index[:160000]
```

```python
fig, ax = plt.subplots(4, 2, figsize=(16, 16))

time_slices = [(i, i+20000) for i in range(0, 160000, 20000)]
for i, (start, end) in enumerate(time_slices):
    row = i // 2
    col = i % 2
    ax[row, col].plot(time[start:end], sen4_smooth[start:end])
    ax[row, col].set_title(f'{start} - {end} millisec')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(4, 2, figsize=(16, 16))

time_slices = [(i, i+20000) for i in range(0, 160000, 20000)]
for i, (start, end) in enumerate(time_slices):
    row = i // 2
    col = i % 2
#     ax[row, col].plot(time[start:end], sen4[start:end])
#     ax[row, col].set_title(f'{start} - {end} millisec')
    
    sen4_detrend = signal.detrend(sen4[start:end])
    n = len(sen4_detrend)
#     f_hat = np.fft.fft(sen4, n)
#     PSD = f_hat*np.conj(f_hat)/n
#     L = np.arange(1,np.floor(n/2), dtype='int')
#     freq = np.arange(n)

#     ax[row, col].plot(np.log(freq[L]),PSD[L], color='c')
#     plt.xlim(0,500)
#     n = len(sen4[start:end])
#     sig = sen4[start:end]
    yf = rfft(sen4_detrend)
    xf = rfftfreq(n,0.001) # 0.001 denotes time in millisec
    
    ax[row,col].plot(xf,np.abs(yf))
    ax[row,col].set_title(f'{start} - {end} millisec')
    
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(3, 3, figsize=(20, 16))

time_slices = [(i, i+20000) for i in range(0, 180000, 20000)]
for i, (start, end) in enumerate(time_slices):
    row = i // 3
    col = i % 3
    widths = np.arange(1, 31)
    cwtmatr, freqs = pywt.cwt(signal.detrend(sen4_smooth[start:end]), widths, 'mexh')
    ax[row,col].imshow(cwtmatr, extent=[-1, 1, 1, 31], cmap='PRGn', aspect='auto',
           vmax=abs(cwtmatr).max(), vmin=-abs(cwtmatr).max()) 

plt.tight_layout()
plt.show()                              
```

```python
fig, ax = plt.subplots(3, 3, figsize=(20,16))

time_slices = [(i, i+20000) for i in range(0, 180000, 20000)]
for i, (start, end) in enumerate(time_slices):
    row = i // 3
    col = i % 3
    widths = np.arange(1, 31)
    cwtmatr, freqs = pywt.cwt(signal.detrend(sen4_smooth[start:end]), widths, 'mexh')
    ax[row,col].imshow(np.absolute(cwtmatr), extent=[-1, 1, 1, 31], cmap='PRGn', aspect='auto',
           vmax=abs(cwtmatr).max(), vmin=-abs(cwtmatr).max()) 

plt.tight_layout()
plt.show()       
```
